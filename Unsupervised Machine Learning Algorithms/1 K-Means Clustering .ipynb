# K-Means Clusterring using Income Spent

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("C:/Users/rushi/OneDrive/Desktop/INTERNSHIP COURSE-Pantech/3 Machine learning-30Days Internship/Unsupervised ML-DayWise/DAY_19_ClusterringUsingIncomeSpent_KMeans/dataset.csv")
```

```python
print(dataset.head())
print(dataset.shape)
```

```python
Income = dataset['INCOME'].values
Spend = dataset['SPEND'].values
```

```python
X = np.array(list(zip(Income,Spend)))
```

```python
X[:10]
```

```python
# Finding the Optimized K Value
```

```python
from sklearn.cluster import KMeans
wcss = []    # wcss = Within-cluster Sum of Square
for i in range(1,11):
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(X)
    wcss.append(km.inertia_)

plt.plot(range(1,11),wcss,color="red", marker="8")
plt.title('Optimal K Value')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

# Fitting the k_means to the dataset with k=4

```python
model = KMeans(n_clusters=4, random_state=0)
y_means = model.fit_predict(X)
```

# Visualizing the clusters for k=4
Cluster 1: Customers with medium income and low spend

Cluster 2: Customers with high income and medium to high spend

Cluster 3: Customers with low income

Cluster 4: Customers with medium income and high spend

```python
plt.scatter(X[y_means==0,0],X[y_means==0,1],s=50, c='brown',label='1')
plt.scatter(X[y_means==1,0],X[y_means==1,1],s=50, c='blue',label='2')
plt.scatter(X[y_means==2,0],X[y_means==2,1],s=50, c='green',label='3')
plt.scatter(X[y_means==3,0],X[y_means==3,1],s=50, c='cyan',label='4')

plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1],s=150,marker='s',c='red', label='Centroids')

plt.title('Income Spent Analysis', c='red')
plt.xlabel('Income',c='blue')
plt.ylabel('Spent',c='blue')
plt.legend()
plt.show()
```
