# Digit Recognition using | RANDOM FOREST

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("C:/Users/rushi/OneDrive/Desktop/Projects DS,ML,DL,AI/GitHub Uploded Resporatory/Pantech/Digit Recognition using RANDOM FOREST/digit.csv")
```

```python
dataset
```

# Segregating Dataset into X & Y

```python
x = dataset.drop('label',axis=1)
x[:10]
```

```python
y = dataset.iloc[:,:1]
y
```

# Splitting Dataset into Train & Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
print('x_train size:',x_train.shape)
print('x_test size:',x_test.shape)

print('y_train size:',y_train.shape)
print('y_test size:',y_test.shape)
```

# Training

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

# Model Accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy of the model:{0}".format(accuracy_score(y_test, y_pred)*100) )
```
